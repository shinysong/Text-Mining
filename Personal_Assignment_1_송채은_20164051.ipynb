# 과제 설명
## 코드 설명
1. 시도 함수 : 17개 시도 url를 넣으면, infobox의 내용이 담긴 df와, 각 시도에 속한 시군구 링크의 list를 return한다.

2. 시군구 함수 : url을 넣으면 시도 함수와 같이 infobox 내용의 df와 각 읍,면,동 링크 list를 return한다.

3. 읍면동 함수 : url을 넣으면 infobox의 내용만 return한다.

4. 메인 함수 : 시도 url만 넣으면, 2-3중 for문으로 데이터 프레임을 완성시킨다.

유의사항 : 하나의 메인함수로 하고 싶었지만, 각 시도, 시군구, 읍면동마다 인포박스의 선택자가 다른경우가 있고 기타 예외처리가 필요한 사항이 많아서, 각 함수들을 조금씩 변형시켜서 사용하였다.

```
def sido(str):
  url = str
  html_text = requests.get(url).text
  soup = bs(html_text, "html.parser")
  
  title = soup.select_one('h1#firstHeading').text.strip()
  print(title,url)
  sido = soup.select_one('#mw-content-text > div.mw-parser-output > table.infobox > tbody')
  sido2 = sido.select('tr')
  i = 0
  for s in sido2: 
    try:
      if (s.select_one('th').text.strip() == "표어"):
        num = i
        break
      elif (s.select_one('th').text.strip() == "나라"):
        num = i
        break
      elif (s.select_one('th').text.strip() == "국가"):
        num = i
        break
      else:
        i += 1
    except:
      i += 1
      continue 

  sido3 = sido2[num:]
  sido3_index_list = []
  sido3_content_list = []



  for s in sido3:
    try:
      if (s.text.strip() == '행정'or s.text.strip() == '인구' or  s.text.strip() == '지리' or  s.text.strip() == '상징'or s.text.strip() == '현황' or s.text.strip() == '역사' or s.text.strip() == '도청' ):
        continue
      elif (s.select('*')[0].text.strip() == '홈페이지'):
        sido3_index_list.append(s.select('*')[0].text.strip())
        sido3_content_list.append(s.select_one('a').get('href'))
      elif (s.select('*')[0].text.strip() == '총인구'):
        sido3_index_list.append('인구')
        sido3_content_list.append(s.select_one('td').text.strip())
      elif (s.select('*')[0].text.strip() == '소재지' or s.select('*')[0].text.strip() == '면사무소' or s.select('*')[0].text.strip() == '시청소재지' or s.select('*')[0].text.strip() == '읍사무소' or s.select('*')[0].text.strip() == '동주민센터' or s.select('*')[0].text.strip() == '동사무소'):
        sido3_index_list.append('주소')
        sido3_content_list.append(s.select_one('td').text.strip())

      else:
        sido3_index_list.append(s.select_one('th').text.strip())
        sido3_content_list.append(s.select_one('td').text.strip())
    except:
      sido3_index_list.append('홈페이지')
      sido3_content_list.append(s.select_one('a').get('href'))


  df2 = pd.DataFrame(data = sido3_content_list, index=sido3_index_list, columns=[title])

  #시군구링크 가져오는 함수
  sigungu_table = soup.find("table",class_="wikitable sortable")
  sigungu_url = sigungu_table.select("b a[href]")
  sigungu_list = []
  for s in sigungu_url:
    sigungu_list.append('https://ko.wikipedia.org'+s.get('href'))

  return df2, sigungu_list
```

```
def dong(str):
  url = str
  html_text = requests.get(url).text
  soup = bs(html_text, "html.parser")
  
  title = soup.select_one('h1#firstHeading').text.strip()

  print(title, url)
  dong = soup.select_one('#mw-content-text > div.mw-parser-output > table.infobox')
  dong2 = dong.select('tbody > tr')
  i = 0
  for d in dong2:
    try:
      if (d.select_one('th').text.strip() == "면적"):
        num = i
      elif (d.select_one('th').text.strip().startswith('인구') == True):
        num = i
      else:  
        i += 1
    except:
      i += 1

  dong3 = dong2[num:]
  dong3_index_list = []
  dong3_content_list = []
  for s in dong3:
    if (s.select_one('th').text.strip() == '웹사이트'):
      dong3_index_list.append('홈페이지')
      dong3_content_list.append(s.select_one('a').get('href'))
    elif (s.select_one('th').text.strip() == '소재지' or s.select_one('th').text.strip() == '면사무소' or s.select_one('th').text.strip() == '시청소재지' or s.select_one('th').text.strip() == '읍사무소' or s.select_one('th').text.strip() == '동주민센터' or s.select_one('th').text.strip() == '동사무소'):
      dong3_index_list.append('주소')
      dong3_content_list.append(s.select_one('td').text.strip())
          
    else:
      
      dong3_index_list.append(s.select_one('th').text.strip().replace(' ',"").replace("•","").replace('\xa0',""))
      dong3_content_list.append(s.select_one('td').text.replace('\xa0',"").strip())
  n = 0
  for i in dong3_index_list:
    if (i.startswith('인구(') == True):
      j = i[:2]
      dong3_index_list[n] = j
    else:
      n+=1
  df3 = pd.DataFrame(data = dong3_content_list, index=dong3_index_list, columns=[title])
  return df3
```

```
def crawling1(str):
  #시도 함수로 서울특별시 url 제공한뒤, 서울시 인포구하고, 서울의 시군구 url 가져오기
  #이중 for문?
  df1, list1 = sido(str)
  for i in list1:
    df2, list2 = sigungu3(i)
    df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
    for j in list2:
      df3 = dong2(j)
      df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
  return df1
```

```
def crawling2(str): # 강원도
  df1, list1 = sido2(str)
  for i in list1:
    if (i == 'https://ko.wikipedia.org/wiki/%ED%8C%8C%EC%9D%BC:Gangwon-map_ko.png'):
      continue
    df2, list2 = sigungu3(i)
    df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
    for j in list2:
      df3 = dong2(j)
      df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
  return df1
```

```
def crawling3(str): # 충청남도
  df1, list1 = sido(str)
  for i in list1:
    df2, list2 = sigungu3(i)
    df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
    for j in list2:
      if (j=="https://ko.wikipedia.org//wiki/%EC%84%9C%EB%B6%81%EA%B5%AC" or j=="https://ko.wikipedia.org//wiki/%EB%8F%99%EB%82%A8%EA%B5%AC"):
        df3, list3 = sigungu3(j)
        df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
      else:
        df3 = dong2(j)
        df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
  return df1
```

```
def crawling4(str): # 충청북도, 전남, 전북
  df1, list1 = sido(str)
  for i in list1:
    if (i == "https://ko.wikipedia.org/wiki/%EB%A7%88%EC%82%B0%EC%8B%9C" or i == "https://ko.wikipedia.org/wiki/%EC%A7%84%ED%95%B4%EA%B5%AC"):
      continue
    df2, list2 = sigungu3(i) 
    df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
    for j in list2: 
    ## 특이사항이 있는 주소 넣어주기
      if (j=="https://ko.wikipedia.org/wiki/%EC%99%84%EC%82%B0%EA%B5%AC" or j=="https://ko.wikipedia.org/wiki/%EB%8D%95%EC%A7%84%EA%B5%AC"
      or j=="https://ko.wikipedia.org/wiki/%EB%A7%88%EC%82%B0%ED%9A%8C%EC%9B%90%EA%B5%AC"or j=="https://ko.wikipedia.org/wiki/%EB%A7%88%EC%82%B0%ED%95%A9%ED%8F%AC%EA%B5%AC"or j=="https://ko.wikipedia.org/wiki/%EC%A7%84%ED%95%B4%EA%B5%AC"):
        df3, list3 = sigungu3(j)
        df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
        for k in list3:
          df3 = dong2(k)
          df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
      else:
        df3 = dong2(j)
        df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
  return df1
```

```
def crawling5(str): #경기도
  df1, list1 = sido(str)
  for i in list1:
    if (i == "https://ko.wikipedia.org/wiki/%EC%88%98%EC%9B%90%EC%8B%9C" or i == "https://ko.wikipedia.org/wiki/%EA%B3%A0%EC%96%91%EC%8B%9C"or i == "https://ko.wikipedia.org/wiki/%EC%9A%A9%EC%9D%B8%EC%8B%9C"
    or i == "https://ko.wikipedia.org/wiki/%EC%84%B1%EB%82%A8%EC%8B%9C"or i == "https://ko.wikipedia.org/wiki/%EC%95%88%EC%82%B0%EC%8B%9C"or i == "https://ko.wikipedia.org/wiki/%EC%95%88%EC%96%91%EC%8B%9C"): # 고양시
      df2, list2 = sigungu3(i)
      df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
      for j in list2:#덕양구
        df2, list3 = sigungu3(j)
        df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
        for k in list3:
          df3 = dong2(k)
          df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
    else:
      df2, list2 = sigungu3(i)
      df1 = pd.merge(df1, df2, how ='outer', left_index=True, right_index=True)
      for j in list2:
        df3 = dong2(j)
        df1 = pd.merge(df1, df3, how ='outer', left_index=True, right_index=True)
  return df1
```
